````python
df3 = pd.read_excel('Records_DataFrame.xlsx')
df3[['Phone Number']]
````

Now, we create a new list variable to store the data we are going to manipulate. We delete the ```-``` separation and replace the NaN values to empty strings, so we can work more comfortable with the phone numbers

````python
fixed_phone = df3['Phone Number'].str.replace('-','')
fixed_phone = fixed_phone.fillna('') 
fixed_phone
````

What we do next is to associate each country or city name to their corresponding country phone code. In order to do that we first associate them with the geographic area where they belong.

We extract the unique countries and cities names 

````python
unique_countries = df3['Country'].unique()
unique_countries
````

Then, we find the geographic areas where our countries and cities belong using the ```geopy``` library

````python
country_list = []

for i in range(len(unique_countries)):
    geolocator = Nominatim(user_agent="google") 
    location = geolocator.geocode(unique_countries[i], language="en")
    country_name = location.address.split(',')[-1].replace(' ','')
    
    country_list.append(country_name)

country_list
````

And we associate both columns in a dictionary 

````python
country_dict = dict(zip(unique_countries,country_list))
country_dict
````

Now, we can use this dictionary as a reference to assign the corresponding country code to each country or city name

````python
for i in range(len(fixed_phone)):
    
    for x in country_dict:
        if df3['Country'][i] == x:
            country = (country_dict[x])
   
    if country == 'Ireland':
        country_code = '(+353)'
        
    elif country == 'UnitedKingdom':
        country_code = '(+44)'
        
    else :
        country_code = ''
````

We assign the country phone code from each country or city name to the ```country_code``` variable

Once we get the country phone code we can add it to the phone number and start modify the phone numbers to the asked format in this exercise.

But first we have to handle the missing values and phone numbers starting with zeros. That's why we do this:

````python
# This keeps ocurring inside the previous for loop

if fixed_phone[i] == '': # Handling missing values
        fixed_phone[i] = 'Missing phone number'
    else :
        while fixed_phone[i][0] == '0': # Phone numbers starting with zeros
            fixed_phone[i] = fixed_phone[i][1:]
        
        # Then we can arranged the phone numbers to the asked format of the exerise
        fixed_phone[i] = country_code + ' ' + fixed_phone[i][:4] + ' ' + fixed_phone[i][4:]
````

The entire for loop would look like this, with its final result:

````python
for i in range(len(fixed_phone)):
    
    for x in country_dict:
        if df3['Country'][i] == x:
            country = (country_dict[x])
   
    if country == 'Ireland':
        country_code = '(+353)'
        
    elif country == 'UnitedKingdom':
        country_code = '(+44)'
        
    else :
        country_code = ''
    
    if fixed_phone[i] == '':
        fixed_phone[i] = 'Missing phone number'
    else :
        while fixed_phone[i][0] == '0':
            fixed_phone[i] = fixed_phone[i][1:]

        fixed_phone[i] = country_code + ' ' + fixed_phone[i][:4] + ' ' + fixed_phone[i][4:]

fixed_phone
````

We finish this exercise by adding this column to our data frame
